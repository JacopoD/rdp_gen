```python
import pandas as pd
x = pd.read_csv("/Users/jdv/Desktop/in_out_with_fits.csv")
```

```python
x['removed_wse_perc'] = x['removed_wse']/x['input']*100
x['removed_fit_perc'] = x['removed_fit']/x['input']*100
x['rest'] = x['output']/x['input']*100
print(sum(x['removed_wse_perc'])/len(x['removed_wse_perc']))
print(sum(x['removed_fit'])/len(x['removed_fit']))
x = x.iloc[::5, :]
```

```python
x
```

```python
import matplotlib.pyplot as plt

fig10 = plt.figure(figsize=(40, 15))
ax10 = plt.subplot()
plt.xticks(fontsize = 40)
plt.yticks(fontsize = 40)
plt.xlabel('xlabel', fontsize=40)
plt.ylabel('ylabel', fontsize=40)

i = x["input"].values
a = x["removed_wse_perc"].values
b = x["removed_fit_perc"].values
c = x["rest"].values

print(len(x["rest"].values))
width = 300
ax10.bar(i, a, width, label="Removed by WSE")
ax10.bar(i, b, width ,bottom=a, label="Removed because of overflow")
ax10.bar(i, c, width ,bottom= b+a, label="Output")
ax10.set_ylabel('%')
ax10.set_xlabel('Number of Samples')
ax10.legend(fontsize=40)
plt.show()

fig10.savefig('in_out_with_fits.pdf', dpi=120)
```
